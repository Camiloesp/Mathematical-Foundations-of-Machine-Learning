````python
import numpy as np
import matplotlib.pyplot as plt
````

````python
# Bring back old formula y = x^2 + 2x + 2

def f(my_x):
    my_y = my_x**2 + 2*my_x + 2
    return my_y
````

````python
x = np.linspace(-10, 10, 1000)
y = f(x)
````

````python
fig, ax = plt.subplots()
plt.axvline(x=0, color='lightgray')
plt.axhline(y=0, color='lightgray')
_=ax.plot(x,y)
````

````python
def diff_demo(my_f, my_x, my_delta):
    return (my_f(my_x + my_delta) - my_f(my_x)) / my_delta
````

````python
deltas = [1, 0.1, 0.01, 0.001, 0.0001, 0.00001, 0.000001]
````

````python
for delta in deltas:
    print(diff_demo(f,2,delta))
````

````python
for delta in deltas:
    print(diff_demo(f,-1,delta))
````
