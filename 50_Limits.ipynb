# Calculus of the Infinitesimals

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(-10,10,1000) # start, finish, n points
```

```python
# if y = x^2 + 2x + 2
y = x**2 + 2*x + 2
```

```python
fig, ax = plt.subplots()
_=ax.plot(x,y)
```

```python
fig, ax = plt.subplots()
ax.set_xlim([-2,0])
ax.set_ylim([0,2])
_=ax.plot(x,y)
```

```python
fig, ax = plt.subplots()
ax.set_xlim([-1.5,-0.5])
ax.set_ylim([0.5,1.5])
_=ax.plot(x,y)
```

```python
fig, ax = plt.subplots()
ax.set_xlim([-1.1,-0.9])
ax.set_ylim([0.9,1.1])
_=ax.plot(x,y)
```
