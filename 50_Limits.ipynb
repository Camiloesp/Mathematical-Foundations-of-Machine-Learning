# Calculating Limits

````python
fig, ax = plt.subplots()
plt.axvline(x=0, color='lightgray')
plt.axhline(y=0, color='lightgray')
ax.set_xlim(-5,10)
ax.set_ylim(-10,80)
plt.axvline(x=5,color='purple',linestyle='--')
plt.axhline(y=37,color='purple',linestyle='--')
_=ax.plot(x,y)
````

```
lim     (x^2-1)/(x-1)
x->1
```

````python
def my_fxn(my_x):
    my_y = (my_x**2-1)/(my_x-1)
    return my_y
````

````python
my_fxn(2)
````

````python
# Divide by 0 error
# my_fxn(1)
````

````python
my_fxn(0.9)
````

````python
my_fxn(0.999)
````

````python
my_fxn(1.1)
````

````python
my_fxn(1.001)
````

````python
y = my_fxn(x)
y
````

````python
fig, ax = plt.subplots()
plt.axvline(x=0, color='lightgray')
plt.axhline(y=0, color='lightgray')
ax.set_xlim(-1,5)
ax.set_ylim(-1,5)
plt.axvline(x=1,color='purple',linestyle='--')
plt.axhline(y=2,color='purple',linestyle='--')
_=ax.plot(x,y)
````

```
lim     (sin x) / x
x->0
```

````python
def sin_fxn(my_x):
    my_y = np.sin(my_x) / my_x
    return my_y
````

````python
# Division by zero error:
# sin_fxn(0)
````

````python
sin_fxn(0.1)
````

````python
sin_fxn(0.001)
````

````python
sin_fxn(-0.1)
````

````python
sin_fxn(-0.001)
````

````python
y = sin_fxn(x)
````

````python
fig, ax = plt.subplots()
plt.axvline(x=0, color='lightgray')
plt.axhline(y=0, color='lightgray')
ax.set_xlim(-10,10)
ax.set_ylim(-1,2)
plt.axvline(x=0,color='purple',linestyle='--')
plt.axhline(y=1,color='purple',linestyle='--')
_=ax.plot(x,y)
````

```
lim             25/x
x->Infinite
```

````python
def inf_fxn(my_x):
    my_y = 25/my_x
    return my_y
````

````python
inf_fxn(1e3)
````

````python
inf_fxn(1e6)
````

````python
y = inf_fxn(x)
````

````python
fig, ax = plt.subplots()
plt.axvline(x=0, color='lightgray')
plt.axhline(y=0, color='lightgray')
ax.set_xlim(-10,10)
ax.set_ylim(-300,300)
plt.axvline(x=0,color='purple',linestyle='--')
plt.axhline(y=1,color='purple',linestyle='--')
_=ax.plot(x,y)
````

````python
left_x = x[x<0]
right_x = x[x>0]
````

````python
left_y = inf_fxn(left_x)
right_y = inf_fxn(right_x)
````

````python
fig, ax = plt.subplots()
plt.axvline(x=0, color='lightgray')
plt.axhline(y=0, color='lightgray')
ax.set_xlim(-10,10)
ax.set_ylim(-300,300)
ax.plot(left_x, left_y, c='C0')
_=ax.plot(right_x,right_y, c='C0')
````
